### Make Results Table

```python
results_df = pd.DataFrame(columns=['method', 'language', 'F1_micro', 'F1_macro', 'accuracy'])

for subfolder in os.listdir('./results_table'):
    for csv_file in os.listdir(f'./results_table/{subfolder}'):
        print(csv_file)
        df = pd.read_csv(f'./results_table/{subfolder}/{csv_file}', sep='\t')
        df['method'] = csv_file
        df = df.reindex(columns=['method', 'language', 'F1_micro', 'F1_macro', 'accuracy'])
        display(df.columns)

        results_df = results_df.append(df, ignore_index=True)
        if len(results_df.columns) > 5:
            print('AHA HERE!')
    
display(results_df)
results_df.to_csv(f'../full_results.csv', index=None)
```

```python
def extract_threshold(string):
    if 'threshold' not in string:
        return 0.5
    else:
        return float(f'0.{string.split("_")[-1].split(".")[0]}')

results_df['threshold'] = results_df['method'].apply(lambda x: extract_threshold(x))
```

```python
def clean_name(string):
    # clean .csv
    csv_ind = string.find('.csv')
    if csv_ind != -1:
        string = string[:csv_ind+1]

    # clean ensemble names
    and_ind = string.find('and')
    if and_ind != -1:
        string = 'ensemble_' + string

    # clean threshold text
    ind = string.find('_threshold')
    if ind == 0:
        return string
    else:
        return string[:ind]

results_df['method'] = results_df['method'].apply(lambda x: clean_name(x))
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

y_scaled = True
use_preset_palette = True

"""
Group Data
"""
results_df = results_df.sort_values(by=["threshold"]) 
grouped = results_df.groupby("language")

"""
Plot each group of data
"""

if use_preset_palette:
    palette = {'xlm_roberta_large': '#e60049', 'ensemble_po_ru_omitted_original_model_xlm-roberta': '#0bb4ff', 'po_ru_omitted_english_model_xlm-roberta': '#50f891', 'po_ru_omitted_original_model_xlm-roberta': '#e6d800', 'ensemble_xlm_roberta_large': '#9b19f5', 'roberta_large_english': '#ffa300'}
else:
    colors = ["#e60049", "#0bb4ff", "#50f891", "#e6d800", "#9b19f5", "#ffa300"]
    palette = {method:color for method,color in zip(results_df["method"].unique(), colors)}

y_min, y_max = min(results_df[['F1_micro', 'F1_macro']].min(numeric_only = True)), max(results_df[['F1_micro', 'F1_macro']].max(numeric_only = True))

fig, axs = plt.subplots(len(grouped)+1, 2, figsize=(12,28))
for i, (name, group) in enumerate(grouped):
    for j, method in enumerate(group["method"].unique()):
        method_group = group[group["method"] == method]
        axs[i, 0].plot(method_group["threshold"], method_group["F1_micro"], label=f"{method}", color=palette[method])
        axs[i, 1].plot(method_group["threshold"], method_group["F1_macro"], label=f"{method}", color=palette[method], linestyle='--')

    axs[i, 0].set_xticks(results_df['threshold'].unique())
    axs[i, 1].set_xticks(results_df['threshold'].unique())
    axs[i, 0].set_xticklabels([])
    axs[i, 1].set_xticklabels([])
    if y_scaled:
        axs[i, 0].set_ylim(y_min, y_max)
        axs[i, 1].set_ylim(y_min, y_max)
    axs[i, 0].set_title(f"{name} (micro)")
    axs[i, 0].set_ylabel("score")
    axs[i, 1].set_title(f"{name} (macro)")
    axs[i, 1].set_ylabel("score")
    # add line
    for threshold in results_df['threshold'].unique():
        axs[i, 0].axvline(x=threshold, color='lightgrey', linestyle='--', linewidth=0.5)
        axs[i, 1].axvline(x=threshold, color='lightgrey', linestyle='--', linewidth=0.5)

"""
Plot average graph
"""
mean_df = results_df.groupby(["method", "threshold"]).mean().reset_index()

# Plot the mean values for each method
for j, method in enumerate(mean_df['method'].unique()):
    method_group = mean_df[mean_df["method"] == method]
    axs[len(grouped), 0].plot(method_group["threshold"], method_group["F1_micro"], label=f"{method}", color=palette[method])
    axs[len(grouped), 1].plot(method_group["threshold"], method_group["F1_macro"], label=f"{method}", color=palette[method], linestyle='--')
for threshold in results_df['threshold'].unique():
    axs[len(grouped), 0].axvline(x=threshold, color='lightgrey', linestyle='--', linewidth=0.5)
    axs[len(grouped), 1].axvline(x=threshold, color='lightgrey', linestyle='--', linewidth=0.5)

if y_scaled:
    axs[len(grouped), 0].set_ylim(y_min, y_max)
    axs[len(grouped), 1].set_ylim(y_min, y_max)
axs[len(grouped), 0].set_title(f"average (micro)")
axs[len(grouped), 0].set_xlabel("confidence_threshold")
axs[len(grouped), 0].set_ylabel("score")
axs[len(grouped), 1].set_title(f"average (macro)")
axs[len(grouped), 1].set_xlabel("confidence_threshold")
axs[len(grouped), 1].set_ylabel("score")
axs[len(grouped), 0].xaxis.set_tick_params(rotation=60)
axs[len(grouped), 1].xaxis.set_tick_params(rotation=60)
axs[len(grouped), 0].set_xticks(results_df['threshold'].unique())
axs[len(grouped), 1].set_xticks(results_df['threshold'].unique())


"""
Finish
"""

# Add a single legend
handles, labels = axs[0, 0].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper right', fontsize=11,)

fig.patch.set_facecolor('white')

plt.savefig(f'../all_results_compared_y_scaled_{y_scaled}.png')
plt.tight_layout()
plt.show()
```
